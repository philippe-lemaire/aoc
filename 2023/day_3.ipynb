```python
from input_data_3 import sample_data, data

import re

n_pattern = re.compile(r"\d+")
symbol_pattern = re.compile(r"[^\.1234567890]")

def contact(span, sspan):
   
    # span is to the left
    if abs(sspan[0] - (span[-1]-1)) <= 1:     
        return True
    # span is to the right
    if abs(span[0] - (sspan[-1]-1)) <= 1:
        return True
    # is there is some overlap
    s1 = set(range(*span))
    s2 = set(range(*sspan))
    if s1.intersection(s2):
        return True
    return False

def get_sum_part_numbers(data=data):
    numbers = []
    symbols = []
    for i, line in enumerate(data.splitlines()):
        n_search= re.finditer(n_pattern, line)
        s_search = re.finditer(symbol_pattern, line)
        for m in  n_search:
            numbers.append((i, m.span(), int(m.group(0))))
        for m in s_search:
            symbols.append((i, m.span(), m.group(0)))
    total = 0
    for n in numbers:
        row = n[0]
        span = n[1]
        value = n[2]
        # check if it "touches" a symbol
        for symbol in symbols:
            srow = symbol[0]
            sspan = symbol[1]
            if abs(srow - row) <= 1 and contact(span, sspan):
                total += value
    return total


get_sum_part_numbers()
```

```python
def contact(span, sspan):
    '''checks if 2 spans are close enough, independently of row number'''
    # span is to the left
    if abs(sspan[0] - (span[-1]-1)) <= 1:     
        return True
    # span is to the right
    if abs(span[0] - (sspan[-1]-1)) <= 1:
        return True
    # is there is some overlap
    s1 = set(range(*span))
    s2 = set(range(*sspan))
    if s1.intersection(s2):
        return True
    return False

def get_gears(data=data):
    stars = []
    numbers = []
    for i, line in enumerate(data.splitlines()):
        # find all numbers
        n_search= re.finditer(re.compile(r"\d+"), line)
        for m in  n_search:
            numbers.append((i, m.span(), int(m.group(0))))
        # find all stars *
        star_search = s_search = re.finditer('\*', line)
        for m in star_search:
            stars.append((i, m.span(), m.group(0)))
    # find gears among stars, compute ratios but only add if it's a gear
    gear_ratio_sum = 0
    for star in stars:
        gear_ratio = 1
        star_row = star[0]
        star_position = star[1]
        counter = 0
        for n in numbers:
            n_row = n[0]
            n_position = n[1]
            n_value = n[2]
            # check if rows and positions are close enough
            if abs(star_row - n_row) <= 1 and contact(star_position, n_position):
                # it's contact
                counter +=1
                if counter <=2:
                    gear_ratio *= n_value
                
        if counter == 2:
            # it's a gear
            gear_ratio_sum += gear_ratio
    return gear_ratio_sum
```
